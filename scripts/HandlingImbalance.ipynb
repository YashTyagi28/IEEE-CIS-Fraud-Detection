```python
import pandas as pd
```

```python
df = pd.read_csv("CleanedData150.csv")
df.head()
```

```python
df["isFraud"].value_counts()
```

```python
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline

X = df.drop("isFraud", axis=1)
y = df["isFraud"]

# Step 1: Undersample majority to 300k
undersample = RandomUnderSampler(sampling_strategy={0: 300000, 1: 20663}, random_state=42)

# Step 2: Oversample minority (fraud) to 200k
smote = SMOTE(sampling_strategy={1: 200000}, random_state=42)

# Pipeline
resample_pipeline = Pipeline([
    ("undersample", undersample),
    ("smote", smote)
])

X_res, y_res = resample_pipeline.fit_resample(X, y)

print("Original:", y.value_counts())
print("Resampled:", y_res.value_counts())
```
